## Тестовый ноут для проверки обращения к модели

Проверка работы модели через api

```python
import yaml # для кредов 
from openai import OpenAI # коннекторы к моделькам
```

```python
# читаем id для подуключения
with open(r'..\test\cred.yml', 'r', encoding='utf-8') as file:
    config = yaml.safe_load(file) 

API_KEY = config['api']
FOLDER_ID = config['folder']
PROMT_ID = config['prompt']
```

```python
# Настройка модели
model = f"gpt://{FOLDER_ID}/yandexgpt/rc"

# Инициализация клиента
client = OpenAI(
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    api_key=API_KEY,
    project=FOLDER_ID
)

# Создание запроса
res = client.responses.create(
    model=model,
    prompt={'id':PROMT_ID},
    input="Какой проходной балл был на эконмику в 2023 году ?"
)
```
